```python
## LOAD INCUBATOR SUMMARIES
##
sheet0_headers = [
           'incubator', 'cash', 'date', 'cohort_id', 'cohort_name', 'cohort_start',
           'cohort_end', 'geo', 'num_interested', '_ignore1', '_ignore2',  'num_applications', 'source_applications',
           'source_applications2',
           'question1', 'question2', 'age_lead_applicants', 'age_winning_founders',
           'percent_male', 'percent_female', 'num_employees', 'percent_teams_one_female',
           'percent_teams_half_female',
           'num_beneficiaries', 'num_beneficiaries_increase_6months', 'num_countries',
           'num_investor_opportunities', 'total_new_investment', 'num_teams_on_icrf',
           'num_teams_on_other', 'num_teams_public_service_contracts', 'num_investors_engaged',
           'withdrawls'
]
def load_sheet0(filename):
    tmp = pandas.read_excel(filename,0,skip_footer=8,names=sheet0_headers)
    tmp = tmp[ tmp['incubator'].notnull() ]
    tmp = tmp.drop(['_ignore1','_ignore2'],axis=1)
    return tmp
# -- 
summary_bethnal = load_sheet0('data/Bethnal Green Ventures (3.2.2014).xlsx')
summary_hub     = load_sheet0('data/Hub Launchpad (3.2.2014).xlsx')
summary_north   = load_sheet0('data/Social Incubator North (3.2.2014).xls')
summary_wayra   = load_sheet0('data/Wayra UnLtd (3.2.2014).xlsx')
summary = pandas.concat([ summary_bethnal, summary_hub, summary_north, summary_wayra ], ignore_index=True)
## -- This doesn't tell us a lot...
summary
```

```python
## IMPORT SUMMARY OF VENTURES
sheet1_headers = ['venture', 'target', 'target2', 'sector', 
 'sector2', 'invested_type', 'invested']
# Pad out the list of columns
actual_colz = 31
fake_colz = ['_drop%d'%i for i in range(actual_colz-len(sheet1_headers)) ]
# --
def load_sheet1(filename):
    tmp = pandas.read_excel(filename, 1, names=sheet1_headers+fake_colz)
    tmp = tmp.drop(fake_colz,axis=1)
    return tmp
ventures_bethnal = load_sheet1('data/Bethnal Green Ventures (3.2.2014).xlsx')
ventures_hub     = load_sheet1('data/Hub Launchpad (3.2.2014).xlsx')
ventures_north   = load_sheet1('data/Social Incubator North (3.2.2014).xls')
ventures_wayra   = load_sheet1('data/Wayra UnLtd (3.2.2014).xlsx')
ventures_bethnal['incubator'] = 'bethnal'
ventures_hub['incubator']     = 'hub'
ventures_north['incubator']   = 'north'
ventures_wayra['incubator']   = 'wayra'
ventures = pandas.concat([ventures_bethnal,ventures_hub,ventures_north,ventures_wayra],ignore_index=True)
ventures.tail()
```

```python
target_list = list(ventures.target.value_counts().index)
# --
merged = {
    'Conservation of the natural environment':'Environmental',
    'Asset Management & Managed Workspace':'Housing and local facilities'
}
icon_map = {
        u'Arts, heritage, sport and faith': 'icon-picture',
        u'Employment, training and education': 'icon-trophy',
        u'Environmental': 'icon-leaf',
        u'Family, friends and relationships': 'icon-group',
        u'Housing and local facilities': 'icon-home',
        u'Income and financial inclusion': 'icon-money',
        u'Mental health and wellbeing': 'icon-medkit',
        u'Physical health': 'icon-heart'
}
sector_list = list(ventures.sector.map(lambda x:merged.get(x,x)).value_counts().index)
def get_icons(df):
    
    counts = df.sector.value_counts()
    for key in sector_list:
        for n in range(counts.get(key,0)):
            deduplicated = merged.get(key,key)
            icon = icon_map.get(deduplicated,'icon-question')
            yield icon
list(get_icons(ventures_hub))
```

```python
# RENDER PIE CHART. Data structure for a chart:
# [{'name':'foo','value':123}, {'name'...} ...]
# -- This can be used to render a pie chart of sectors or targets
def pandas_pie(frame, column, reindex):
    tmp = frame.fillna(0).groupby([column])\
        .venture.count()\
        .reindex(index=reindex,fill_value=0)
    return [ {'name':x,'value':y} for (x,y) in tmp.iteritems() ]
pandas_pie(ventures_hub, 'sector', sector_list)
```

```python
## Data export
def icrf_dict(x):
    return {'name':x[1][0], 'amount':x[1][1], 'url':x[1][3]}
def sib_dict(name,df,img,url):
    summary_rows = summary[summary['incubator']==name]
    mean = df.invested.mean()
    withdrawls = summary_rows.withdrawls.sum()
    funding = summary_rows.cash.sum()
    to_int = lambda x: -1 if numpy.isnan(x) else int(x)
    return {
            'name':name,
            'total_funding':to_int(funding),
            'mean_investment':to_int(mean),
            'withdrawls':to_int(withdrawls),
            'investment_sectors':list(get_icons(df)),
            'img' : img, 
            'url' : url,
            'sector_pie':pandas_pie(df,'sector',sector_list),
            'target_pie':pandas_pie(df,'target',target_list),
            }
            
data = {
    'icrf_mean':icrf.cash.mean().round(),
    'icrf_count':int(icrf.cash.count()),
    'icrf_items':[ icrf_dict(x) for x in icrf.iterrows() ],
    'sib': [
        sib_dict( 'Bethnal Green Ventures', ventures_bethnal, '/images/logo_bethnal.png', 'http://bethnalgreenventures.com/' ),
        sib_dict( 'Hub Launchpad', ventures_hub, '/images/logo_hub.png', 'http://www.hublaunchpad.net/' ),
        sib_dict( 'Social Incubator North', ventures_north, '/images/logo_socialincubatornorth.png', 'http://socialincubatornorth.org.uk/' ),
        sib_dict( 'Wayra UnLtd', ventures_wayra, '/images/logo_wayra.png', 'http://wayra.org/unltd/' ),
    ],
    'icon_to_sector' : dict([(v,k) for k,v in icon_map.iteritems()]),
}
with open('tmp.json','w') as f:
    import json
    json.dump(data,f)
data
```
